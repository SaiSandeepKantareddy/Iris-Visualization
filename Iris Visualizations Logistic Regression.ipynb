```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
#X=[[1,2,3,4,5,6],[1,2,3,4,5,6,]]
#Y=[1,2,3,4]
```

```python
df=pd.read_csv("Iris.csv")
df.drop(['Id'],axis=1,inplace=True)
```

```python
X=df.iloc[0:,0:4]
Y=df.iloc[0:,4:]
```

```python
df['Species'] = df['Species'].astype('category')
df['target'] = pd.Categorical(df['Species']).codes
df['target'].value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
sn.pairplot(data=df,hue="Species")
```

```python
sn.swarmplot(x="Species",y="PetalLengthCm",data=df)
```

```python
sn.boxplot(x="Species",y="PetalLengthCm",data=df)
```

```python
sn.violinplot(x="Species",y="PetalLengthCm",data=df)
```

```python
model=linear_model.LogisticRegression()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
score=accuracy_score(predictions,y_test)
print("The accuracy score is",score)
```
